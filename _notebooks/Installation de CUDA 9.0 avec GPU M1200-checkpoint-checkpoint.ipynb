Set up symlinks for gcc/g++

```python
sudo ln -s /usr/bin/gcc-6 /usr/local/cuda/bin/gcc
sudo ln -s /usr/bin/g++-6 /usr/local/cuda/bin/g++
```

setup your paths

```python
echo 'export PATH=/usr/local/cuda-9.0/bin:$PATH' >> ~/.bashrc
echo 'export LD_LIBRARY_PATH=/usr/local/cuda-9.0/lib64:$LD_LIBRARY_PATH' >> ~/.bashrc
source ~/.bashrc
```

# Installation CUDNN

In order to install cuDNN v7.2, you have to download it, first register here https://developer.nvidia.com/developer-program/signup

Then download cuDNN v7.2 from https://developer.nvidia.com/cudnn

Choose the following package:

```python
tar -xzvf cudnn-9.0-linux-x64-v7.2.1.38.tgz
```

copy the following files into the cuda toolkit directory

```python
sudo cp -P cuda/include/cudnn.h /usr/local/cuda-9.0/include
sudo cp -P cuda/lib64/libcudnn* /usr/local/cuda-9.0/lib64/
sudo chmod a+r /usr/local/cuda-9.0/lib64/libcudnn*
```

Finally, to verify the installation, check

# Installation de l'environnement virtuel

Installation de pip:

```python
wget https://bootstrap.pypa.io/get-pip.py
sudo python3 get-pip.py
sudo pip install virtualenv virtualenvwrapper
sudo rm -rf ~/.cache/pip get-pip.py
```

On ouvre ensuite le bashrc, et on colle les lignes suivant la commande sur la fin

```python
sudo nano  ~/.bashrc
    export WORKON_HOME=$HOME/.virtualenvs
    export VIRTUALENVWRAPPER_PYTHON=/usr/bin/python3
    source /usr/local/bin/virtualenvwrapper.sh
```

enregister le fichier : ctrl + X puis O p

```python
source ~/.bashrc
```

On peut maintenant créer l'environnement

```python
mkvirtualenv nom_environnement -p python3
```

Pour effacer un virtualenv, il suffit de taper la commande suivante:

```python
rmvirtualenv nom_environnement
```
